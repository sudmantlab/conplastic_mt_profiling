```python
ref_seq_file = "/Users/isabelserrano/Documents/Science/Analyses/Conplastic_Strains/files_and_analyses/input_files/chrM.fa"
ref_seq = SeqIO.read(ref_seq_file, "fasta")
```

```python
#this allows us to print the seq record as a fasta file (and it looks like the fasta file we have)
print(ref_seq.format("fasta"))
```

```python
#we want to remove the D-Loop region in the fasta file so that we don't have these coordinate as possible
#positions for mutations in our simulator 

#remember that it's end coordinate exclusive
chrM_wo_dloop = ref_seq[:15423]
```

```python
#adding appropriate descriptors to our seq record for our fasta file
chrM_wo_dloop.id = "chrM_wo_dloop"
chrM_wo_dloop.name = "chrM_wo_dloop"
chrM_wo_dloop.description = "chrM_wo_dloop"
```

```python
#testing to see how the fasta file will look 
print(chrM_wo_dloop.format("fasta"))
```

```python
#exporting the fasta file for chrM without the D-Loop
outdir = "/Users/isabelserrano/Documents/Science/Analyses/Conplastic_Strains/files_and_analyses/genomewide_selection_scan/files/"

SeqIO.write(chrM_wo_dloop, outdir + "" +"chrM_wo_dloop.fa", "fasta")
```
